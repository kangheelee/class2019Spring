```python
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
print(wordlist[:10])
```

```python
result = [w for w in wordlist if re.search('^..j..t..$', w)] # ^는 첫번째 글자, .은 한글자, $는 마지막 글자
print(result[:10])
```

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)] #{4}는 4번 반복하는것.  '^[0-9]{4}$'는 숫자가 네번 반복되는 것
print(wordlist[:10])
```

## 숙제 그 아래부분 findall이나 FreqDist 다 한번 해보고!! 기능 다 알아보기!!
수업시간에는 안다뤘지만 시험에 나올 수 있다는 말씀인 듯

```python
word = 'supercalifragilisticexpialidocious'
result = re.findall(r'[aeiou]', word)
print(result)
result = re.findall('[aeiou](..)[aeiou]', word)
print(result)
result = re.findall('[^aeiou].+[^aeiou]', word)
print(result)
result = re.findall('[^aeiou].+?[^aeiou]', word)
print(result)
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cfd = nltk.ConditionalFreqDist(cvs) # CFD works pairwise only
cfd.tabulate()
```

```python
cv_word_pairs = [(cv, w) for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cv_index = nltk.Index(cv_word_pairs)
cv_index['po']
```

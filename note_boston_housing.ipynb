```python
# description
# https://www.kaggle.com/c/boston-housing

# TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu')) # Dense 뒤에 있는 숫자 : 히든 레이어 10개
                                                            #input은 입력 똥그라미 개수, 즉 x값은 13개 
model.add(Dense(1))  # 이전에 input 갯수 정해져있으니까 여기서는 히든레이어만 정의해준 것임, 다음 레이어 한쌍의 부분을 나타냄('add'이니까)
model.compile(optimizer='rmsprop', loss='mse') # 이거는 설명 자세히 안해주심. 카테고리컬에서는 adam, entropy? 써준다고함.
model.summary()
# param은 파라미터인데 화살표의 개수이다. 즉 (x의 개수+1) 곱하기 히든 레이어 개수
# 시험!! 140이라는 수와 11이라는 수가 어떻게 나오는지 나올 듯. 화살표를 만들 때는 1을 꼭 한개 더해줘야한다.
#히든  layer는 relue라는 activation만 넣으면 되는데 # target layer 같은 경우는 목적에 맞는 activation 설정을 해줘야함. -> linear이 기본 값
# 화살표 숫자는 파라미터의 개수! **시험나옴!! : 13 * 10 = 130인데 왜 첫쌍 결과가 140이냐. 13+1(상수)해야하기 때문!
# dense_2에서 11개의 파라미터인 이유는 hidden layer {10개 +1(상수)} * 1(target layer 노드 개수)
```

```python
 model.fit(x_train, y_train, epochs=1000)  # 위에 식이 (1*13) x (13*10) x (10*1)이었는데 epochs=1000해주면 맨 처음 1을 1000으로 바꿔줌
                                           # 1*13 : 한집에 대한 13가지 정보 입력  # 1000*13 : 1000개의 집의 13가지 정보를 얻는 것
#그래서 (None, 13)이나 (None, 1) 뭐 이런식으로 생각하면 됩니다.  # 지금 우리가 하는 것은 "Keras"라는 것.model.fit batch-size
```

```python
# INPUT::
# 1. crim: per capita crime rate by town.
# 2. zn: proportion of residential land zoned for lots over 25,000 sq.ft.
# 3. indus: proportion of non-retail business acres per town.
# 4. chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
# 5. nox: nitrogen oxides concentration (parts per 10 million).
# 6. rm: average number of rooms per dwelling.
# 7. age: proportion of owner-occupied units built prior to 1940.
# 8. dis: weighted mean of distances to five Boston employment centres.
# 9. rad: index of accessibility to radial highways.
# 10. tax: full-value property-tax rate per $10,000.
# 11. ptratio: pupil-teacher ratio by town.
# 12. black: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
# 13. lstat: lower status of the population (percent).

# TARGET::
# medv: median value of owner-occupied homes in $1000s.

model.predict(np.array([0, 1, 50, 1, 0, 5, .5, 10, 1000, 100, 10, 100, 1]).reshape(1,13))
```
